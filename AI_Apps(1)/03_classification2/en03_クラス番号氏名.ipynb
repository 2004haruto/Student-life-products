## 3. アヤメの花を品種に分類（LinearSVC）

### 3-1. 学習用データの準備（共通）

```python
# CSVファイルの読み込み
# pandasライブラリのインポート
import pandas as pd

# csvファイルの読み込み(ファイル名:iris-dataset.csv)
df = pd.read_csv('iris-dataset.csv')
```

```python
# 入力と出力の取り出し
# 特徴量(入力) とアヤメの品種(出力) に分ける
x = df[['がく片長さ', 'がく片幅', '花弁長さ', '花弁幅']]
t = df['種類']
print(x)
print(t)
```

```python
# 訓練用データとテスト用データに分割
# 関数を使うためにインポート
from sklearn.model_selection import train_test_split
# データの分割
x_train, x_test, y_train, y_test = train_test_split(
	x, 
	t, 
	test_size=0.3, 
	random_state=2230033,
	stratify=t,  # 各クラスのデータの割合を保って分割する
)
```

```python
# 表示して確認してもいい
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

### 3-2. モデルの構築（LinearSVC）

```python
# モデルを使用するためにインポート
from sklearn.svm import LinearSVC
```

```python
# モデルの作成
model_LSVC = LinearSVC()
```

```python
# fitに学習用データを入れて学習
model_LSVC.fit(x_train, y_train)
```

```python
# 精度の確認（訓練用とテスト用それぞれ）
score_train = (model_LSVC.score(x_train, y_train))
score_test = (model_LSVC.score(x_test, y_test))

print(score_train)
print(score_test)
```
