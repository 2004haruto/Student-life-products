# 3. 物体検出のサンプル

```python
# ライブラリのインポート
from ultralytics import YOLO
```

## 3-1. モデルの読み込み

```python
# モデルの読み込み
# model = YOLO("モデルファイルパス")
model = YOLO("yolo11n.pt")
```

## 3-2. 検出の実行

```python
# 画像ファイルを指定して検出
results = model.predict("zidane.jpg") # results = model.predict(source="zidane.jpg")
```

```python
# 結果の画像も出力
results = model.predict("zidane.jpg", save=True)
```
