# 1. OpenCVを使用した顔検出

## 1-2. 顔検出プログラムの作成

### 1-2-4. 顔検出プログラムの確認

◇ 必要なライブラリをインポート

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
import cv2
```

◇ Haar-cascade分類器を読み込んでモデル作成

```python
# Haar-cascade分類器を読み込む
cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)
```

◇ 画像を読み込んでグレースケール変換

```python
# 画像を読み込み、グレースケールに変換する
img = cv2.imread("baby.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 元画像かグレースケール画像を表示して確認してもよい
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.show()
```

◇ Haar-cascade分類器を用いて顔を検出する

```python
# Haar-cascade分類器を用いて顔を検出する
face_list = cascade.detectMultiScale(img_gray,minSize=(150,150))
```

◇ 顔が検出されなかった場合はプログラムを終了する

◇ 検出された顔の数だけ処理する

```python
# 検出された顔の数だけ処理する
for(x,y,w,h) in face_list:
    # 顔の座標を出力する
    print("顔の座標=",x,y,w,h)
    # 赤色の線で顔の周りに四角形を描画する
    red = (0,0,255)
    cv2.rectangle(img , (x,y) , (x+w , y+h) , red , thickness=20)
```

```python
# 処理された画像を保存する
cv2.imwrite("face-detect.png",img)
```

```python
# 画像を表示して確認する
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.show()
```

## □ 2-3. モザイクをかける関数の活用

```python
import matplotlib.pyplot as plt  # matplotlibからpyplotをインポート
import cv2  # OpenCVをインポート
from mosaic import mosaic # 自作のmosaic.py内のmosaic関数をインポート

img = cv2.imread("cat.png")  # 画像を読み込み
# まずは表示して確認
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # matplotlibで画像を表示するためにRGBに変換してからimshowで表示
plt.show()  # 表示

```

```python
# モザイクをかける関数
mos = mosaic(img,(100,100,600,450),10)  # mosaic関数を用いてモザイク処理を行う mosaic(画像,(始点x軸 , 始点y軸 , 終点x軸 , 終点y軸))

cv2.imwrite("cat_mosaic.png",mos)  # モザイク処理を施した画像を保存
plt.imshow(cv2.cvtColor(mos,cv2.COLOR_BGR2RGB))  # matplotlibで画像を表示するためにRGBに変換してからimshowで表示
plt.show()  # 表示
```
